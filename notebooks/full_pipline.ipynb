```python
import pandas as pd
import seaborn as sns
from pathlib import Path
```

```python
base_path = Path('/Users/anuj/Documents/CodeBase/Bioinformatics_projects/biological-age-prediction/data/raw')
```

```python
data = {}
for years in base_path.iterdir():
    year = years.name
    
    for features in years.glob('*.xpt'):
        
        feature_name = features.stem.split("_")
        if feature_name[0] == 'P':
            feature_name = feature_name[1]
        else:
            feature_name = feature_name[0]
        
                           
        df = pd.read_sas(features, format='xport', encoding="latin1")
        
        df['year'] = year
        data[(year, feature_name)] = df
        
```

```python
for key, value in data.items():
    print(f'key = {key} & value = {value.shape}')
```
